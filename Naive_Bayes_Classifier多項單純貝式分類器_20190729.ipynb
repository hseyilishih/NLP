```python
#這個分類器適用於離散數據
# MultinomialNB()這裡可以使用平滑參數alpha
```

```python
iris = datasets.load_iris()  #load進iris的資料庫
```

```python
features = iris.feature_names
X = iris.data  
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)  

clf = MultinomialNB()
clf.fit(X_train, y_train) 
#MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True) 
#alpha->smoothing parameter(0 for no smoothing)

y_result = clf.predict(X_test)
accuracy_score(y_test, y_result)

print("Number of mislabeled points out of a total %d points : %d"% (iris.data.shape[0],(y_test != y_result).sum()))
```

```python
def preprocessing(text):
    words = word_tokenize(text)
    #去掉'!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    tokens = [w for w in words if w.lower() not in string.punctuation]
    
    #取得stopwords
    stopw = stopwords.words('english')
    
    found_in_stopwords = [token for token in tokens if token in stopw]
 
        

    #去掉stopwords Call 
    tokens = [token for token in tokens if token not in stopw]    
    
    #取掉 remove words less than three letters
    tokens = [word for word in tokens if len(word)>=3]
    tokens_before = tokens
 
    
    #詞形還原lemmatize
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word) for word in tokens]
    tokens_after = tokens

    
    preprocessed_text = ' '.join(tokens)
    
    
    len_words_text = len(text.split())
    len_words_preprocessed_text = len(preprocessed_text.split()) 

    
 
    ratio = SequenceMatcher(None, text, preprocessed_text).ratio()

    return preprocessed_text 
```

```python
def func_preprocessing(text):
    words = word_tokenize(text)
    #去掉'!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    tokens = [w for w in words if w.lower() not in string.punctuation]
    
    #取得stopwords
    stopw = stopwords.words('english')
    
    found_in_stopwords = [token for token in tokens if token in stopw]
    if len(found_in_stopwords)>0 :
        print('\n 這些字被 stop 剔除了!....',found_in_stopwords)
        

    #去掉stopwords Call 
    tokens = [token for token in tokens if token not in stopw]    
    
    #取掉 remove words less than three letters
    tokens = [word for word in tokens if len(word)>=3]
    tokens_before = tokens
 
    
    #詞形還原lemmatize
    lemma = WordNetLemmatizer()
    tokens = [lemma.lemmatize(word) for word in tokens]
    tokens_after = tokens

    
    preprocessed_text = ' '.join(tokens)
    
    print('compare text 被刪減字元數長度:', len(text) - len(preprocessed_text))
    
    len_words_text = len(text.split())
    len_words_preprocessed_text = len(preprocessed_text.split()) 
    print('compare text 被刪減字元  長度(split方法):', len_words_text - len_words_preprocessed_text)
    
    print('compare text 被刪減字元  長度(re方法):', len(re.findall(r'\w+', text)) - len(re.findall(r'\w+', preprocessed_text)))
    ratio = SequenceMatcher(None, text, preprocessed_text).ratio()
    print('ratio of SequenceMatcher after/before=', ratio )     

    print('\n string before : \n', text)
    print('\n string after: \n', preprocessed_text)
    print('\n compare string before/after difference=', set(text) - set(preprocessed_text))
   
    print('\n tokens before lemma: \n', tokens_before)
    print('\n tokens after lemma: \n', tokens_after)
    print('\n compare tokens before/after difference=', set(tokens_before) - set(tokens_after))
    
    
    #奇怪 11只列出1 (當作兩個1)    
    return preprocessed_text 
```

## <mark style=background-color:grey> Convert a collection of text documents to a Tf-idf matrix </mark> 

```python
stopwords = [word.strip().lower() for word in open("C:/Users/USER/NLP_JERRYWU/data/ENstopwords.txt")]
print('stopwords 總共有:', len(stopwords))
print(*stopwords)
```

### <mark style=background-color:grey> TfidfVectorizer </mark> 

```python
vectorizer = TfidfVectorizer(min_df=0.0025, max_df=0.1, ngram_range=(1, 2),
                         stop_words=stopwords,strip_accents='unicode',
                         sublinear_tf=True,norm='l2')

#min_df忽略文檔頻率過低的權重
#max_df忽略文檔頻率過高的權重（像是停用詞庫）
#ngram_range提取範圍
#刪除重音（accents）字眼
#sublinear_tf線性的tf計算
#Norm標準化詞向量
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['Text_p'], df['Class'], 
                                                    test_size=0.3, random_state= 0)
```

```python
y_train
```

```python
df['Text_p']
```

```python
X_train_v = vectorizer.fit_transform(X_train)
X_test_v = vectorizer.transform(X_test)
```

```python
#在所有的詞上面建立TFIDF的權重，可以去掌握權重
weights = np.asarray(X_train_v.mean(axis=0)).ravel().tolist()
weights_df = pd.DataFrame({'term': vectorizer.get_feature_names(), 'weight': weights})
weights_df.sort_values(by='weight', ascending=False).head(20)
```

## Naive Bayes Classifier

```python

model = MultinomialNB().fit(X_train_v, y_train)
```

```python
y_pred = model.predict(X_test_v)
print(y_pred)

#試試情緒分類？
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
