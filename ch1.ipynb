# Freq

```python
x = '石獅寺有四十四隻石獅子，四十四隻石獅子吃四十四枝濕紫柿子。'
```

```python
#詞頻
def word_freq(x):
    x_dict = dict()
    for text in x:
        #get函數返回指定鍵的值，如果值不在字典中返回默認值
        x_dict[text]=x_dict.get(text,0)+1
    return x_dict

text_list = [text for text in x]

print(text_list)
text_freqdict = word_freq(x)
print(text_freqdict)
```

```python
#排序
from operator import itemgetter
text_sorted=sorted(text_freqdict.items(), key=itemgetter(1), reverse=True)
print(text_sorted)
```

```python
#特定詞頻
text_sorted_conditions = list()

for (text,num) in text_sorted:
    if num > 2:
        text_sorted_conditions.append((text,num)) 

print(text_sorted_conditions)
```

# N-Gram

```python
def bigram(x):
    return [x[i:i+2] for i in range(0,len(x)-1)]

def trigram(x):
    return [x[i:i+3] for i in range(0,len(x)-2)]

bigram_x = bigram(x)
trigram_x = trigram(x)

bigram_x,trigram_x
```

```python
def bigram_freq(x):
    bigram_dict=dict()
    for (c1,c2) in x:
        bigram_dict[(c1,c2)]=bigram_dict.get((c1,c2),0)+1
    return bigram_dict
```

```python
bigram_freq_data = bigram_freq(bigram_x)
```

```python
def trigram_freq(x):
    bigram_dict=dict()
    for (c1,c2,c3) in x:
        bigram_dict[(c1,c2,c3)]=bigram_dict.get((c1,c2,c3),0)+1
    return bigram_dict
```

```python
trigram_freq_data = trigram_freq(trigram_x)
```

```python
bigram_freq_sorted = sorted(bigram_freq_data.items(), key=itemgetter(1), reverse=True)
```

```python
trigram_freq_sorted = sorted(trigram_freq_data.items(), key=itemgetter(1), reverse=True)
```

```python
#透過最大匹配算法（MaxMatch）
import nltk
from nltk.corpus import words  
wordlist = set(words.words())   
def max_match(text):  
    pos2 = len(text)  
    result = ''  
    while len(text) > 0:         
        word = text[0:pos2]
        if word in wordlist:  
            result = result + text[0:pos2] + ' '  
            text = text[pos2:]  
            pos2 = len(text)  
        else:  
            pos2 = pos2-1                 
    return result[0:-1]  
string = 'theyarebirds'  
print(max_match(string))
```
